```python
import itertools
import numpy as np
import math
```

```python
light = ["q1", "q2", "q3", "q4", "q5", "q6"]
ws = list(itertools.combinations(light,2)) + [("vl")]
```

```python
redundant_combs = list(itertools.combinations_with_replacement(ws, 4))
```

```python
clean_combs = {}
ind = 0
for line in redundant_combs:
    counter = {}
    good = True
    for tup in line:
#         if tup[1][0] in counter or tup[1][1] in counter:
        if tup[0] in counter or tup[1] in counter:
            good = False
            break
        else:
#             counter[tup[1][0]] = 1
#             counter[tup[1][1]] = 1        
            counter[tup[0]] = 1
            counter[tup[1]] = 1        
    if good:
        clean_combs[ind] = line
    ind += 1
```

```python
result = [y for x in clean_combs.values() for y in list(itertools.permutations(x))]
```
